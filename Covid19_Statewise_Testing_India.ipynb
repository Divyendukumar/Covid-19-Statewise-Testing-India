```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/My Drive/StatewiseTestingDetails.csv')
df.head()

#Cleaning Data
df = df.fillna(0)

#Removing missing rows
df = df.dropna()

#Analyzing Data (Total sample by each state)
statewise_tests = df.groupby('State')['TotalSamples'].max().sort_values(ascending=False)
statewise_tests.head(10)

#Visualizing Data (Top 5 states by Total test)
import matplotlib.pyplot as plt

top5 = statewise_tests.head(5)
top5.plot(kind='bar', color='orange')

plt.title('Top 5 States by COVID-19 Tests Conducted')
plt.ylabel('Total Samples Tested')
plt.xlabel('State')
plt.show()
```
